```python
X_unnormalized=np.genfromtxt('./ass1_data/data/q1/linearX.csv')
Y=np.genfromtxt('./ass1_data/data/q1/linearY.csv')
```

```python
X=(X_unnormalized-X_unnormalized.mean())/X_unnormalized.std()
X=np.vstack((X, np.ones(X.shape))).T
X
```

```python
def grad(theta, X, Y):
    '''Function to compute partial differentiation wrt theta'''
    add=[0,0]
#     print(X.shape[0])
    for i in range(X.shape[0]):   
        Xi= X.iloc[i,:]
        Yi= Y.iloc[i,:]
        add += Xi*(Yi - theta.dot(Xi)).values[0]
#     print(add)
    return (-1/X.shape[0])*add
```

```python
def LinearRegressionSGD(X,Y, r=0.01, max_iter=2000):
    theta = np.zeros((2,))    
    theta_list=[]
    for i in tqdm(range(max_iter)):
        theta_next= theta - r * np.array(grad(theta, X, Y))
        theta_list.append(theta)
        theta=theta_next
    return theta_list
```

```python
theta_lst = LinearRegressionSGD(X, Y)
```
