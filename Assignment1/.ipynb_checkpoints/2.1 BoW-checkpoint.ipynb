## Assignment 2.1

* In this section we will perform t-SNE visulisation on our amazon fine food reviews dataset(Sample size of 10k due to unsufficient computation power) using BagOfWords Vectorizer

```python
#Loading amazon Dataset and filtering them 
#and mapping reviews to positive or megative values

import sqlite3
import pandas as pd
import numpy as np
import nltk
import string
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
from nltk.stem.porter import PorterStemmer
import re
```

```python
#Loading dataset

con = sqlite3.connect('./database.sqlite') 

#Here we are not getting the reviews with rating 3
filtered_data = pd.read_sql_query("""SELECT * FROM Reviews
WHERE Score != 3""", con) 

def partition(x):
    if x < 3:
        return 'negative'
    return 'positive'

#changing reviews with score less than 3 to be positive and vice-versa
actualScore = filtered_data['Score']
positiveNegative = actualScore.map(partition) 
filtered_data['Score'] = positiveNegative

print(filtered_data.shape)
```

## Now preforming data pre-processing

```python
sorted_data=filtered_data.sort_values('ProductId', axis=0,\
                                      ascending=True,\
                                      inplace=False,\
                                      kind='quicksort',\
                                      na_position='last')
```

Now performing de-deuplication for various subclasses as listed below

* (UserId, ProfileName, Time, Text) At the same thime for same user same reviews shouldn't be there multiple times
* (UserId, ProductId, Time) At the same thime for same user and product same reviews shouldn't be there multiple times
* (UserId, Time) One user can't have multiple reviews at the same timestamp 


```python
dedup1=sorted_data.drop_duplicates(\
                        subset={"UserId","ProfileName","Time","Text"},\
                        keep='first', inplace=False)
print("Getting shape {} after 1st deduplication operation".format(dedup1.shape))

dedup2=dedup1.drop_duplicates(\
                        subset={"UserId","ProductId","Time"},\
                        keep='first', inplace=False)
print("Getting shape {} after 2nd deduplication operation".format(dedup2.shape))

final=dedup2.drop_duplicates(subset={"UserId", "Time"}, keep='first', inplace=False)
print("Getting shape {} after 3rd deduplication operation".format(final.shape))

per=(final['Id'].size*1.0)/(filtered_data['Id'].size*1.0)*100

print("After performing deduplication we have only {}% data left".\
     format(per))
```

Now helpfulnessNumerator can't be greater then HelpfulnessDenominator so removing such entries which have helpfulnessNumerator>helpfulnessDenominator

```python
final=final[final.HelpfulnessNumerator<=final.HelpfulnessDenominator]
```

```python
print(final.shape)
#We removed two entries which had helpfulnessNumerator>HelpfulnessDenominator!!!

print(final['Score'].value_counts())
pcount=len(final[final.Score=='positive'])
ncount=len(final[final.Score=='negative'])
print("Positive : {} %".format(100*pcount/len(final.index)))
print("Negative : {} %".format(100*ncount/len(final.index)))
```

### So after deduplication and removing erroneous entries we have 328730 reviews in our dataset which contains 83.9% positive and 16.1% negative reviews

## Now performing text preprocessing on reviews text
#### <ul><li>Stop-Word Removal, Stemming, Lemmatization</li></ul>

```python
#import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer

stop = set(stopwords.words('english')) #set of stopwords
sno = nltk.stem.SnowballStemmer('english') #initialising the snowball stemmer

def cleanhtml(sentence): #function to clean the word of any html-tags
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, ' ', sentence)
    return cleantext

def cleanpunc(sentence): #function to clean the word of any punctuation or special characters
    cleaned = re.sub('[?|!|\'|"|#]','',sentence)
    cleaned = re.sub('[.|,|)|(|\|/]',' ',cleaned)
    return  cleaned
```

```python
### Removing stopwords, html tags, and punctuation marks from review text 

i=0
str1=' '
final_string=[]
all_positive_words=[] # store words from +ve reviews here
all_negative_words=[] # store words from -ve reviews here.
s=''
for sent in final['Text'].values:
    filtered_sentence=[]
    #print(sent);
    sent=cleanhtml(sent) # remove HTMl tags
    for w in sent.split():
        for cleaned_words in cleanpunc(w).split():
            if((cleaned_words.isalpha()) & (len(cleaned_words)>2)):    
                if(cleaned_words.lower() not in stop):
                    s=(sno.stem(cleaned_words.lower())).encode('utf8')
                    filtered_sentence.append(s)
                    if (final['Score'].values)[i] == 'positive': 
                        all_positive_words.append(s) #list of all words used to describe positive reviews
                    if(final['Score'].values)[i] == 'negative':
                        all_negative_words.append(s) #list of all words used to describe negative reviews reviews
                else:
                    continue
            else:
                continue 
    #print(filtered_sentence)
    str1 = b" ".join(filtered_sentence) #final string of cleaned words
    #print("***********************************************************************")
    
    final_string.append(str1)
    i+=1
```

```python
## Final format of review text after peforming above operations

final_string[0]
```

```python
#final=final.drop('Text', axis=1)
final['Text']=final_string

final.head(5) #Why these reviews are of books instead of food
```

```python
#Storing in in final.sqlite

final.sample()
conn = sqlite3.connect('final.sqlite')
c=conn.cursor()
conn.text_factory = str
final.to_sql('Reviews', conn,  schema=None, if_exists='replace', index=True, index_label=None, chunksize=None, dtype=None)
```

## Performing BoW and then performing t-SNE

```python
# Getting the sample datset form the sqlite file which we stored before

con = sqlite3.connect('./final10k.sqlite') 

final = pd.read_sql_query("SELECT * FROM Reviews", con) 

count_vect = CountVectorizer() #in scikit-learn
final_counts = count_vect.fit_transform(final['Text'].values)
```

```python
#Sanity check

final_counts.get_shape()
```

```python
#Standardizing data before applying t-SNE

from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler(with_mean=False).fit_transform(final_counts)
print(type(standardized_data))
standardized_data_dense=standardized_data.toarray()
print(type(standardized_data_dense))
standardized_data_dense[0]
```

```python
from MulticoreTSNE import MulticoreTSNE as TSNE #importing multicore t-SNE for faster execution

model = TSNE(n_jobs=4, n_components=2, random_state=0)
# default perplexity = 30
# default Maximum number of iterations for the optimization = 1000

tsne_data = model.fit_transform(standardized_data_dense)
tsne_label = final['Score']

# creating a new data frame which help us in ploting the result data
tsne_data = np.vstack((tsne_data.T, tsne_label)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "Score"))

# Ploting the result of tsne
sns.FacetGrid(tsne_df, hue="Score", size=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.title("t-SNE on BoW with perplexity=30 ane n_iter=1000")
plt.show()
```
